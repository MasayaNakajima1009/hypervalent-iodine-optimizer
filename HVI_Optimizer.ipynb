Hypervalent Iodine MLP Geometry Optimization Demo

This Colab notebook performs geometry optimization of hypervalent iodine compounds using a MACE-based machine-learning potential.

Input: XYZ structure
Output: Optimized XYZ structure

```python
## 2. Upload input XYZ and model file

from google.colab import files

print("Upload an input XYZ file")
uploaded_xyz = files.upload()
INPUT_XYZ = list(uploaded_xyz.keys())[0]

print("Upload a trained MACE model file")
uploaded_model = files.upload()
MODEL_PATH = list(uploaded_model.keys())[0]
```

```python
## 3. Geometry optimization

from pathlib import Path
from ase.io import read, write
from ase.optimize import LBFGS
from mace.calculators import MACECalculator

# ===== Settings =====
F_MAX = 0.01
MAX_STEPS = 500
DEVICE = "cuda"   # use "cpu" if GPU is not available

output_name = Path(INPUT_XYZ).stem + "_optimized.xyz"


# ===== Read structure =====
atoms = read(INPUT_XYZ)


# ===== Load MACE calculator =====
calc = MACECalculator(
    model_paths=MODEL_PATH,
    device=DEVICE
)

atoms.calc = calc


# ===== Geometry optimization =====
opt = LBFGS(atoms)
opt.run(fmax=F_MAX, steps=MAX_STEPS)


# ===== Save optimized structure =====
write(output_name, atoms)

print("Optimization finished.")
print("Optimized XYZ:", output_name)
```

```python
## 4. Download optimized XYZ
files.download(output_name)
```
